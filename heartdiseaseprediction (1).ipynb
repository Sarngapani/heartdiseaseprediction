**Importing the dependencies**


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data collection and processing**

```python
#loading csv data in a panda dataframe
heart_data= pd.read_csv("/content/Heart_Disease_Prediction.csv")
```

0--> healthy heart
1--> defective heart

Splitting the features and heart disease

```python
X= heart_data.drop(columns='Heart Disease',axis=1)
Y= heart_data['Heart Disease']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training


Logistic Regression

```python
model= LogisticRegression()
```

```python
#training the lpgistic regression model with trainikng data
model.fit(X_train,Y_train)
```

Model Evaluation  

Accuracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```

Building a Predictive System

```python
input_data= (63,0,4,150,407,0,2,154,0,4,2,3,7)
#change the input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)
# Reshape the numpy arrya as we are predicting for only one instance
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)
prediction= model.predict(input_data_reshape)
print(prediction)
if(prediction[0]=='Absence'):
  print("The person doesn't have a Heart Disease")
else:
  print("The person have a Heart Disease")
```
